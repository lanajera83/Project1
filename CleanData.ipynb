```python
#import variables and read cvs
import pandas as pd
import numpy as np
file = "listings_mexico.csv"
df = pd.read_csv(file, encoding="ISO-8859-1")
df.columns
```

```python
#reduce table for usefull columns
table_df = df.loc[:, ["id", "name", "host_id", "host_name", "neighbourhood_cleansed", "neighbourhood","zipcode", "latitude", "longitude","room_type", "price", "minimum_nights", "number_of_reviews", "last_review", "reviews_per_month", "calculated_host_listings_count", "availability_365", "bed_type", "amenities", "review_scores_rating"]]
table_df.head()
```

```python
#Check data types
table_df.dtypes
```

```python
#Filter database to keep number of reviews higher than 1
#Clean odd characters
table_reviews = table_df.loc[table_df["number_of_reviews"]>=0, ["id", "name", "host_id", "host_name", "neighbourhood_cleansed", "neighbourhood", "zipcode", "latitude", "longitude","room_type", "price", "minimum_nights", "number_of_reviews", "last_review", "reviews_per_month", "calculated_host_listings_count", "availability_365", "bed_type", "amenities", "review_scores_rating"]]
table_reviews['neighbourhood_cleansed'] = table_reviews['neighbourhood_cleansed'].str.replace('Ã©', 'e')
table_reviews['neighbourhood_cleansed'] = table_reviews['neighbourhood_cleansed'].str.replace('Ã¡', 'a')
table_reviews['neighbourhood_cleansed'] = table_reviews['neighbourhood_cleansed'].str.replace('Ã', 'A')
table_reviews['neighbourhood_cleansed'] = table_reviews['neighbourhood_cleansed'].str.replace('Ã³', 'o')

table_reviews['neighbourhood'] = table_reviews['neighbourhood'].str.replace('Ã©', 'e')
table_reviews['neighbourhood'] = table_reviews['neighbourhood'].str.replace('Ã¡', 'a')
table_reviews['neighbourhood'] = table_reviews['neighbourhood'].str.replace('Ã', 'A')
table_reviews['neighbourhood'] = table_reviews['neighbourhood'].str.replace('Ã³', 'o')

table_reviews.head()
```

```python
no_null_table_reviews = table_reviews.dropna(how='any')
```
